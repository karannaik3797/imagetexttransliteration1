```
result_files = [os.path.join("./result",file_) for file_ in os.listdir("./result") if "mask" not in file_ and file_[-3:]!="txt"]
```

```
def read_image(path):
  img = cv2.imread(path)
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  return img
```

```
sample_results = random.sample(result_files, 4)
img0 = read_image(sample_results[0])
img1 = read_image(sample_results[1])
img2 = read_image(sample_results[2])
img3 = read_image(sample_results[3])

fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(15,15))
ax[0,0].imshow(img0)
ax[0,1].imshow(img1)
ax[1,0].imshow(img2)
ax[1,1].imshow(img3)
plt.show()

```
